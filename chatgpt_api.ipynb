# 計算月薪起訖

```python
# ...existing code...
import pandas as pd
import re

df = pd.read_csv("updated_engineer_job_details.csv")

def parse_salary(s):
    s = str(s)
    # 年薪
    if s.startswith("年薪"):
        match = re.search(r'(\d{1,3}(?:,\d{3})*|\d+)[^~\d]*(?:~|－|–|—|至)(\d{1,3}(?:,\d{3})*|\d+)', s)
        if match:
            start = int(match.group(1).replace(',', '')) // 12
            end = int(match.group(2).replace(',', '')) // 12
            return start, end
        match = re.search(r'(\d{1,3}(?:,\d{3})*|\d+)', s)
        if match:
            val = int(match.group(1).replace(',', '')) // 12
            return val, val
        return None, None
    # 月薪
    match = re.search(r'(\d{1,3}(?:,\d{3})*|\d+)[^~\d]*(?:~|－|–|—|至)(\d{1,3}(?:,\d{3})*|\d+)', s)
    if match:
        start = int(match.group(1).replace(',', ''))
        end = int(match.group(2).replace(',', ''))
        return start, end
    match = re.search(r'(\d{1,3}(?:,\d{3})*|\d+)', s)
    if match:
        val = int(match.group(1).replace(',', ''))
        return val, val
    return None, None

df[['Salary start', 'Salary end']] = df['Salary'].apply(lambda x: pd.Series(parse_salary(x)))

df.to_csv("updated_engineer_job_details_with_salary_new.csv", index=False, encoding="utf-8-sig")
print(df[['Salary', 'Salary start', 'Salary end']].tail())
# ...existing code...
```

```python
import pandas as pd

# 假設 df 是你原本的 DataFrame

bonus_cols = ["年終獎金", "業績獎勵金","績效獎金", "考核獎金", "久任獎金", "三節禮金"]
allowance_cols = ["生日禮金", "結婚禮金", "生育禮金", "奠儀", "旅遊代金"]

def merge_bonus_allowance(row):
    bonus_part = "\n".join(
        f"{col}: {str(row[col]) if pd.notna(row[col]) else '無'}" for col in bonus_cols
    )
    allowance_part = "\n".join(
        f"{col}: {str(row[col]) if pd.notna(row[col]) else '無'}" for col in allowance_cols
    )
    return f"{bonus_part}\n---------------------------------------------\n津貼\n{allowance_part}"

df["其他薪資"] = df.apply(merge_bonus_allowance, axis=1)

# 把「其他薪資」插到 Description 和 Welfare 中間
cols = df.columns.tolist()
cols.remove("其他薪資")
desc_idx = cols.index("Description")
cols.insert(desc_idx + 1, "其他薪資")

df = df[cols]

print(df.head())

df.to_csv("updated_engineer_job_details_with_salary_new_order.csv", index=False, encoding="utf-8-sig")
```
